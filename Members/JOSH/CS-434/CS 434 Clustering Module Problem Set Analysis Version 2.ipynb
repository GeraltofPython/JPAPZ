```python
import numpy as np
data1 = np.load('cdata1.npy')
data2 = np.load('cdata2.npy')
seedData = np.load('seed_data.npy')
print(data1)
```

```python
# Problem 1 2D Scatter plot of cdata1

import matplotlib.pyplot as plt

fig, (ax) = plt.subplots(1, 1)
fig.set_size_inches(10, 5)

plt.scatter(data1[:,0], data1[:,1])
plt.show()
```

```python
# Problem 2 KMeans
from sklearn.cluster import KMeans as km
from sklearn.metrics import silhouette_score

clusterSize = 2
clustering = km(n_clusters = clusterSize)
cluster_labels = clustering.fit_predict(data1)

fig, (ax) = plt.subplots(1, 1)
fig.set_size_inches(10, 5)

# Problem 3 & 4 Silhouette Score
clusterCenters = clustering.cluster_centers_
ax.scatter(data1[:, 0], data1[:, 1])
ax.scatter(clusterCenters[:, 0], clusterCenters[:, 1], color = '', s = 50)

for i, j in enumerate(clusterCenters):
    ax.scatter(j[0], j[1], marker='$%d$' % i, color = 'black', s = 100)
    
plt.show()

print("Cluster Size = ", clusterSize)
print("Silhouette score:", silhouette_score(data1, cluster_labels))
    
```
