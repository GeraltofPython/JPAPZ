```python
# Problem 1

import numpy as np
import matplotlib.pyplot as plt
ctrain = np.load('ctrain.npy')
# print(ctrain)

plt.scatter(ctrain[:,0], ctrain[:,1], c = ctrain[:,2])
plt.show()

'''
I can visualize the class boundary being around the y = 0.5
The upper classes seem seperable but there does seem like there is some overlap
'''
```

```python
# Problem 2
from sklearn.naive_bayes import GaussianNB
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs

xTrain = ctrain[:,[0,1]]
yTrain = ctrain[:,2]
model = GaussianNB()
model.fit(xTrain, yTrain)

rng = np.random.RandomState(0)
xTrain2 = [1.5, - 1] + [-10, 3] * rng.rand(300, 2)
yTrain2 = model.predict(xTrain2)

plt.scatter(xTrain[:, 0], xTrain[:, 1], c = yTrain, s = 50, cmap = 'RdBu')
lim = plt.axis()
plt.scatter(xTrain2[:, 0], xTrain2[:, 1], c = yTrain2, s = 20, cmap = 'RdBu', alpha = 0.5)
plt.axis(lim);
```

```python
import sklearn.metrics as met

#graphCurve1 = met.roc_curve(ctrainX, ctrainY)
#graphCurve2 = met.roc_curve(ctrainX, ctrainY)


score = model.predict(xTrain)
graphCurve1 = met.roc_curve(yTrain, score)
graphCurve2 = met.roc_curve(yTrain2, score)

print(graphCurve1)
print(graphCurve2)

plt.plot(graphCurve1, graphCurve2)
plt.show()
```

```python
# Problem 12 for seed_data

seedData = np.load('seed_data.npy')
xSeed = seedData[:,[0, 1, 2, 3, 4, 5, 6]]
ySeed = seedData[:, 7]

print(xSeed.shape)
```

```python
seedModel = GaussianNB()
seedModel.fit(xSeed, ySeed)
ySeed2 = seedModel.predict(xSeed)

print(classification_report(ySeed, ySeed2))
```

```python
seedN3 = KNeighborsClassifier(n_neighbors = 3)
seedN3 = seedN3.fit(xSeed, ySeed)
seedN3Pred = seedN3.predict(xSeed)

print('\033[1m' + "Test: k = 1")
print('\033[0m')
print(classification_report(ySeed, seedN3Pred))

seedN7 = KNeighborsClassifier(n_neighbors = 7)
seedN7 = seedN7.fit(xSeed, ySeed)
seedN7Pred = seedN7.predict(xSeed)

print('\n\033[1m' + "Test: k = 3")
print('\033[0m')
print(classification_report(ySeed, seedN7Pred))
```

```python
seedDepth3 = tree.DecisionTreeClassifier(max_depth = 3)
seedDepth3 = seedDepth3.fit(xSeed, ySeed)
newSeedDepth3 = seedDepth3.predict(xSeed)

print('\033[1m' + "Tree Depth: 3" + '\033[0m')
print(classification_report(ySeed, newSeedDepth3))


seedDepth2 = tree.DecisionTreeClassifier(max_depth = 2)
seedDepth2 = seedDepth2.fit(xSeed, ySeed)
newSeedDepth2 = seedDepth2.predict(xSeed)

print('\033[1m' + "Tree Depth: 2" + '\033[0m')
print(classification_report(ySeed, newSeedDepth2))
```
